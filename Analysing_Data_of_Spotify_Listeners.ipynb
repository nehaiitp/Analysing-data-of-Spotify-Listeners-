```python
import matplotlib.pyplot as plt

# 1️⃣ Subscription type distribution (Bar Chart)
plt.figure(figsize=(6,4))
df["subscription_type"].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Distribution of Subscription Types")
plt.xlabel("Subscription Type")
plt.ylabel("Number of Listeners")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2️⃣ Average minutes listened by gender (Bar Chart)
plt.figure(figsize=(6,4))
df.groupby("gender")["minutes_listened"].mean().plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Average Minutes Listened by Gender")
plt.xlabel("Gender")
plt.ylabel("Avg Minutes Listened")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 3️⃣ Churned vs Active Users (Pie Chart)
plt.figure(figsize=(6,6))
df["churned"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=["lightcoral","lightblue"],
    labels=["Active","Churned"]
)
plt.title("Churned vs Active Listeners")
plt.ylabel("")  # remove y-axis label
plt.show()
```
